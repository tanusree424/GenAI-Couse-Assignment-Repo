# Task  1

## Write sales record to a file

```python
sales = [1200,450,980,1500,3000]
with open("sales_data.txt","w") as files:
    for amount in sales:
        files.write(str(amount)+ "\n")
    files.close()

# Reopen the file and print contents
file = open("sales_data.txt", "r")
content = file.read()
print("File Contents:")
print(content)

file = open("sales_data.txt", "w")

for i in range(len(sales)):
    if i == len(sales) - 1:
        file.write(str(sales[i]))
    else:
        file.write(str(sales[i]) + ",")

file.close()



```

# Task 4

## Generate Summary Report From sales file

```python
# sales files data convert into integer
with open("sales_data.txt","r") as file:
   lines= file.readlines()
   for line in lines:
      print(int(line))
   file.close()

# Calculate Total Sales

total_sales =0
with open("sales_data.txt","r") as file:
   lines = file.readlines()
   for line in lines:
      sale = int(line.strip())
      total_sales += sale
   print(f"Total Sales: {total_sales}")
# Calclute Highest Number Sales
with open("sales_data.txt","r") as file:
   lines = file.readlines()
   list_sale_item =[]
   for line in lines:
      sale = int(line.strip())
      list_sale_item.append(sale)
   max_sale = max(list_sale_item)
   low_sale = min(list_sale_item)
   avarage_sale  = total_sales / len(list_sale_item)
   print(f"Maximum Sales is: {max_sale}")
   print(f"Minimlum or lowest Sales is: {low_sale}")
   print(f"Average Sale is {avarage_sale}")


      

     
```

## Create Product into file (user input) 

```python
products = {}
def taking_user_input():
    
    for i in range(3):
        
        product_name = input(f"Enter Product {i+1} Name")
        product_price = float(input("Enter product Price:"))
        products[product_name] =product_price
    return products
taking_user_input()
# print("All Products",all_products)
## Product.txt in proper formatting
with open("products.txt","w") as file:
    for index, (key, value) in enumerate(products.items(), start=1):
        file.write(f"{index}. {key} | {value}\n")

    print(f"Products Added Successfully to the file:{file.name}")
    file.close()
## Read the File
with open("products.txt","r") as file:
    print(f"Print with proper formatting:")
    for line in file:
        line = line.strip()
        if line:
            name , price = line.split("|")
            name = name.strip()
            price = float(price)
        print(f" \n {name} | {price}")

```
