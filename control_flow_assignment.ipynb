```python
order = input("Enter order amount: ")

# Check if input is valid integer (including negative check if needed)
if order.startswith('-') and order[1:].isdigit():
    order_amount = int(order)
elif order.isdigit():
    order_amount = int(order)
else:
    print("Error: Please enter a valid numeric value.")
    exit()

# Apply discount rules
if order_amount >= 2000:
    discount = order_amount * 0.15
elif order_amount >= 1500:
    discount = order_amount * 0.10
elif order_amount >= 1000:
    discount = order_amount * 0.07
else:
    discount = 0

discount = round(discount, 2)
final_amount = round(order_amount - discount, 2)

print("Original Amount:", order_amount)
print("Discount:", discount)
print("Final Amount:", final_amount)

# optional add 5% tax

tax = round(final_amount * 0.05,2)
total = final_amount + tax
print(f"Subtotal after discount: {final_amount}")
print(f"Tax Amount: {tax}")
print(f"Final Total: {total}")


```

# Task 2

```python
# Order List
order_list = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discount_count = 0

print("Order  |  Discount  |  Final Amount")
print("--------------------------------------")

for order_amount in order_list:
    
    if order_amount >= 2000:
        discount = order_amount * 0.15
    elif order_amount >= 1500:
        discount = order_amount * 0.10
    elif order_amount >= 1000:
        discount = order_amount * 0.07
    else:
        discount = 0

    discount = round(discount, 2)
    final_amount = round(order_amount - discount, 2)

    total_revenue += final_amount

    if discount > 0:
        discount_count += 1

    #  Print outside discount condition
    print(order_amount, " | ", discount, " | ", final_amount)

print("--------------------------------------")
print("Total Revenue After Discount:", round(total_revenue,2))
print("Number of Orders with Discount:", discount_count)
```

# Task 3

```python
# Order List
order_list = [1200, 2500, 800, 1750, 3000]

# While loop taking input from user and prees 'q' to exit he loop

while True:
    user_input = input(
        "\nPress 1 to Add Order\n"
        "Press 2 to Show All Orders and Total\n"
        "Press q to Exit\n"
        "Enter choice: "
    )

    if user_input not in ["1", "2", "q"]:
        print("Invalid input. Please try again.")
    
    # Press 1  to Enter a Valid Order Amount Added to Order List
    elif user_input == "1":
        amount = input("Enter Order Amount: ")
        
        if amount.isdigit():
            order_list.append(int(amount))
            print("Order added successfully.")
        else:
            print("Invalid amount. Must be a number.")
        ''' Press 2  for Showing Order Amount and check the amount as it is eligile for discount or not 
        If eligible then calculate the  Discount then Show the discount amont for that particular order amount 
        and also show the discount amount and final amount also total revenue as well
        '''

    elif user_input == "2":
        total_revenue = 0
        
        print("\nOrder | Discount | Final Amount")
        print("----------------------------------")
        
        for order_amount in order_list:
            
            if order_amount >= 2000:
                discount = order_amount * 0.15
            elif order_amount >= 1500:
                discount = order_amount * 0.10
            elif order_amount >= 1000:
                discount = order_amount * 0.07
            else:
                discount = 0

            discount = round(discount, 2)
            final_amount = round(order_amount - discount, 2)
            
            total_revenue += final_amount
            
            print(order_amount, "|", discount, "|", final_amount)

        print("----------------------------------")
        print("Total Revenue:", round(total_revenue, 2))
    
    # press 'q' button to exit the loop 
    
    elif user_input == "q":
        print("Thank you!")
        break
```

```python
# Daily Sales Report
daily_sales = [200, 150, 0, -1, 50, 300]

total_sales = 0
''' using for loop if there is -1  then loop will be brak and if value is 
 0 then loop will be continue last ly shown the total sales  '''
for day_sales in daily_sales:

    if day_sales == -1:
        print("Corrupted Data")
        break

    elif day_sales == 0:
        print("Today there is no sales")
        continue

    total_sales += day_sales
    print(f"Day Sales: {day_sales} | Total Sale: {total_sales}")

print("-----------")
print(f"Final Total Sale is: {total_sales}")
```
