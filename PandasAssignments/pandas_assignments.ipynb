# Task 2

## Mathematical Operations On series

```python
import pandas as pd

marks = pd.Series([70, 80, 65, 90, 75])
print("Original Marks:")
print(marks)

add_grace = marks + 5
print("\nAfter Adding 5 Grace Marks:")
print(add_grace)

subtract_marks = marks - 2
print("\nAfter Subtracting 2 Marks:")
print(subtract_marks)

multiply_marks = marks * 1.05
print("\nAfter Multiplying by 1.05:")
print(multiply_marks)

divide_marks = marks / 2
print("\nAfter Dividing by 2:")
print(divide_marks)

```

# Task 3

### Python functionalities on Series

```python
marks = pd.Series([70, 80, 65, 90, 75])
maximum_marks = max(marks)
print("Maximum Marks:", maximum_marks)
minimum_marks = min(marks)
print("Minimum Marks:", minimum_marks)
total_marks = sum(marks)
print("Total Marks:", total_marks)
mean_marks = marks.mean()
print("Mean Marks:", mean_marks)
result = marks.apply(lambda marks: "Pass" if marks >= 70 else "Fail")
print("\nPass/Fail Result:")
print(result, "\n")

total_pass = (marks >= 70).sum()
print("\nTotal Pass:", total_pass)
```

# Task 5

## Important DataFrame Function

```python
import pandas as pd
students = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Pooja"],
    "Marks": [78, 85, 90, 66, 72],
    "Subjects": ["Math", "Math", "Science", "Science", "Math"]
}
students_df = pd.DataFrame(students)
print("Information about DataFrame:")
print(students_df.info())
print("\nSummary Statistics:")
print(students_df.describe())
print("Top Records:")
print(students_df.head())
print("\nBottom Records:")
print(students_df.tail())
print("\nSorted DataFrame by Marks (Descending):")
print(students_df.sort_values(by="Marks", ascending=False))
print("\nReset Index after Sorting:")
students_df_sorted = students_df.sort_values(by="Marks", ascending=False)
students_df_sorted = students_df_sorted.reset_index(drop=True)
print(students_df_sorted)
```

# Task 6

##  Filtering and Conditional Selection

```python
import pandas as pd
students = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Pooja"],
    "Marks": [78, 85, 90, 66, 72],
    "Subjects": ["Math", "Math", "Science", "Science", "Math"]
}

students_df = pd.DataFrame(students)
print("Original DataFrame:")
print(students_df)
print("Scores greater than 75:")
print(students_df[students_df["Marks"] > 75])
print("\nStudents enrolled in Math:")
print(students_df[students_df["Subjects"] == "Math"])
avarage_marks = students_df["Marks"].mean()
print("\nAverage Marks:", avarage_marks)
print("\nStudents with Marks above Average:")
print(students_df[students_df["Marks"] > avarage_marks])
print("\nFailed Students:")
print(students_df[students_df["Marks"] < 70])
```

# Task 7

## Grouping Basic Analysis

```python
import pandas as pd
students = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha", "Pooja"],
    "Marks": [78, 85, 90, 66, 72],
    "Subjects": ["Math", "Math", "Science", "Science", "Math"]
}
students_df = pd.DataFrame(students) 
## Find Average per subject using groupby
average_per_subject = students_df.groupby("Subjects")["Marks"].mean()
print("Average Marks per Subject:")
print(round(average_per_subject, 2))
## Count of students per subject
count_per_subject = students_df["Subjects"].value_counts()
print("\nCount of Students per Subject:")
print(count_per_subject)
## Maximum marks per subject
max_marks_per_subject = students_df.groupby("Subjects")["Marks"].max()
print("\nMaximum Marks per Subject:")
print(max_marks_per_subject)
```

## Mini Use Case for Sales Data Analysis

```python
import pandas as  pd
sales = {
    "Day":['Mon', 'Tue','Wed','Thu','Fri'],
    "Revenue":[1200,1500,900,2000,1800]

}
sales_df = pd.DataFrame(sales)
print("Sales DataFrame:", sales_df)
print("Total Revenue:", sales_df["Revenue"].sum())
print("Average Daily Revenue:", sales_df["Revenue"].mean())
print("Highest Revenue Day:", sales_df.loc[sales_df["Revenue"].idxmax(), "Day"])
print("Days where Revenue is above Average:")
print(sales_df[sales_df["Revenue"] > sales_df["Revenue"].mean()]["Day"])
print("Plot Revenue over Days:")
sales_df.plot(x="Day", y="Revenue", kind="bar")
```
