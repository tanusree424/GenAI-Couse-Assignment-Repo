## Safe Division Utility

```python
numerator = input("Enter numerator value here:...")
denominator =input("Enter Denomenator Value here:..")
try:
    numerator_value = int(numerator)
    denominator_value = int(denominator)

    
except ValueError:
    print("Value Must be an Integer")
except  ZeroDivisionError:
    print("Zero Cannot e divided")
else:
    result = numerator_value / denominator_value
    print("The Result of the division is:", result)
finally:
    print("Operation Completed")
```

# Task 2

### Bill Calculator with  Error Handling

```python
prices  = [120 , 350 , 'abc',500 ,-200 , 80]
total =0
for p in prices:
  try:
        if not isinstance(p,(int, float)):
            raise TypeError
        if p < 0:
            raise ValueError("Negative Value cannot be allowed")
        total += p
  except TypeError:
      print("Value is not a number")
      continue
  except ValueError as e:
      print(e)
      continue
  else:
      print("Running Total is", total)
    
  finally :
      print("Work Done") 
      
    


```

# Task 3

## Custom Validation

```python
def check_age(age):
    if age < 1 or age > 120:
        raise ValueError("Age must be between 1 and 120")
    
    else:
        print("Age is", age)

def main():
    age = input("Enter Age: ")
    try:
        age_value = int(age)
        check_age(age_value)
    except ValueError as e:
        print("Error:", e)
    

main()
```

## File Reader with exception handling

```python
filename = input("Enter the file name: ")

try:
    with open(filename, "r") as file:
        for i in range(3):
            print(file.readline().strip())

except FileNotFoundError:
    print("File Not Found")
```

## Mini Programe: save shopping cart

```python
prices=[]

    
while True:
    crp = input("Enter Prices here or to quit press q")
    if crp == "q":
        break
    try:
        price = float(crp)
        prices.append(price)
    except ValueError:
         print("Price Must be an integer")
   
    
print("Final Carts",prices) 
print("Total Number of Cart", len(prices))
print("Total Amount of Price:", sum(prices))




```
