## Creating Numpy Arrays

```python
import numpy as np
arr1 = np.arange(1,10)
arr2 = np.arange(1,10).reshape(3,3)
arr3 = np.array([10,20,30,40,50])
print("arr1:",arr1, "Shape:",arr1.shape,"Data Type:",arr1.dtype)
print("arr2:",arr2,"Shape:",arr2.shape,"Data Type:",arr2.dtype)
print("arry_list:",arr3,"Shape:",arr3.shape,"Data Type:",arr3.dtype)
```

## Important Mathematical Operations

```python
A = np.array([10,20,30,40])
B = np.array([1,2,3,4])

##  Result for Addition
addition = np.sum(A+B)
## Subtraction 
subtraction = np.subtract(A,B)

## Multiplication
multiplication = np.prod(A*B)
multiplication
## Division
division =  np.divide(A,B)
## Power
power = np.power(A,B)
print("Addition of A & B is:", addition)
print("Subtraction of A & B is:", subtraction)
print("Multiplication of A & B is:", multiplication)
print("Power of A & B is:", power)
```
