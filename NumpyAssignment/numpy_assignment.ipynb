## Creating NumpyArrays

```python
import numpy as np
arr1 = np.arange(1, 10)
arr2 =np.arange(1,10).reshape(3,3)
arr3 = np.array([10,20,30,40,50])

print("array 1:\n",arr1, "\nshape:", arr1.shape , "datatype:", arr1.dtype)
print("array 2:\n",arr2, "\nshape:", arr2.shape , "datatype:", arr2.dtype)
print("array 3:\n",arr3, "\nshape:", arr3.shape , "datatype:", arr3.dtype)
```

## Task 2

## Important mathematical Operations

```python
A = np.array([10,20,30,40])
B = np.array([1,2,3,4])
print("Addition:", A + B)
print("Subtraction:", A - B)    
print("Multiplication:", A * B)
print("Division:", A / B)
print("Power:", A ** 2)
```

## Task 3

## Important Numpy Mathematical Formulas

```python
values  = np.array([2,4,6,8,10])
print("Square root:", np.sqrt(values))
print("Exponential:", np.exp(values))
print("Logarithm:", np.log(values))
print("sum of all elements:", np.sum(values))
print("Cumulative sum of elements:", np.cumsum(values))
```

# Task 4

## Aggregation Operations

```python
data = np.array([[10,20,30],[40,50,60],[70,80,90]])
print("Row wise sum:", np.sum(data, axis=1))
print("Column wise sum:", np.sum(data, axis=0))
print("Minimum value:", np.min(data))
print("Maximum value:", np.max(data))
print("Mean value:", np.mean(data))
print("Minimum value along rows:", np.min(data, axis=1))
print("Maximum value along rows:", np.max(data, axis=1))
print("Max value along columns:", np.max(data, axis=0))
print("Minimum value along columns:", np.min(data, axis=0))
```

# Task 5

## Statistical Operations (Core Focus)

```python
marks = np.array([78,85,90,66,72,88,95,60])
print("Mean:", np.mean(marks))
print("Median:", np.median(marks))
print("Variance:", np.var(marks))
print("Standard Deviation:", np.std(marks))
print("Minimum:", np.min(marks))
print("Maximum:", np.max(marks))  
print("range:" , np.max(marks) - np.min(marks))  
```

# Task 6

## Percentile & Sorting

```python
marks = np.array([78,85,90,66,72,88,95,60])
print("Sorted marks:", np.sort(marks))
## Find 25th percentile
print("25th percentile:", np.percentile(marks, 25))
## Find 50th percentile (median)
print("50th percentile (median):", np.percentile(marks, 50))
## Find 75th percentile
print("75th percentile:", np.percentile(marks, 75))
## Count how many student  score are above avarage
average_score = np.mean(marks)
above_average_count = np.sum(marks > average_score)
print("Number of students scoring above average:", above_average_count)
```

# Task 7

## Mini Use Cse Sales Analysis:

```python
sales = np.array([1200, 1500,900 ,2000,1800, 1700, 1600])
# Total weekly sales
total_sales = np.sum(sales)
print("Total weekly sales:", total_sales)
# Average daily sales
average_sales = np.mean(sales)
print("Average daily sales:", average_sales)
#Highest and lowest sales day
highest_sales = np.max(sales)
lowest_sales = np.min(sales)
print("Highest sales in a day:", highest_sales)
print("Lowest sales in a day:", lowest_sales)
# Standard deviation of sales
sales_std_dev = np.std(sales)
print("Standard deviation of sales:", sales_std_dev)
## Indentify the sales where sales are above average
above_average_sales = sales[sales > average_sales]
print("Sales above average:", above_average_sales)
```
