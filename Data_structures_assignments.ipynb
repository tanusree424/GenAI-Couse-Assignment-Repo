# Task 3

```python
# Task 3: Product Pricing

# 1. Create dictionary
price_dict = {
    "Mobile": 20000,
    "Smartwatch": 5000,
    "Laptop": 55000,
    "Books": 800,
    "Shoes": 2500,
    "Socks": 300
}

# 2. Add new product
price_dict["Microwave"] = 12000
print("After Adding Microwave:", price_dict)

# Update price of existing product
price_dict["Laptop"] = 60000
print("After Updating Laptop Price:", price_dict)

# Remove product (handle if not exists)
product_to_remove = "Tablet"

if product_to_remove in price_dict:
    del price_dict[product_to_remove]
    print(f"{product_to_remove} removed successfully.")
else:
    print(f"{product_to_remove} does not exist in dictionary.")

# 3. Print average price
total_price = sum(price_dict.values())
average_price = round(total_price / len(price_dict),2)

print("Average Price:", average_price)

# Extra: Product with max and min price
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)

print("Product with Maximum Price:", max_product, price_dict[max_product])
print("Product with Minimum Price:", min_product, price_dict[min_product])
```

# Task 4

```python
# Task 4: Combined Operations

products = ["Mobile","Smartwatch","Laptop","Books","Shoes","Socks"]

price_dict = {
    "Mobile": 20000,
    "Smartwatch": 5000,
    "Laptop": 60000,
    "Books": 800,
    "Shoes": 2500,
    "Socks": 300
}

categories = [
    "Electronics",
    "Electronics",
    "Electronics",
    "Books",
    "Footwear",
    "Clothing"
]

# 1. Create catalog list of tuples
catalog = []

for i in range(len(products)):
    product_name = products[i]
    price = price_dict[product_name]
    category = categories[i]
    catalog.append((product_name, price, category))

print("Catalog:", catalog)

# 2. Create category_to_products dictionary
category_to_products = {}

for item in catalog:
    name, price, category = item
    
    if category not in category_to_products:
        category_to_products[category] = []
    
    category_to_products[category].append(name)

print("Category to Products:", category_to_products)

# 3. Category with maximum number of products
max_category = max(category_to_products, key=lambda k: len(category_to_products[k]))

print("Category with Maximum Products:", max_category)
print("Products in that Category:", category_to_products[max_category])
```
