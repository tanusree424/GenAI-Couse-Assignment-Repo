# Task 1

```python
# Task 1

def apply_discount(price, discount_percent=5):

    # Extra condition: discount should not exceed 60%
    if discount_percent > 60:
        print("Discount cannot exceed 60%. Setting to 60%.")
        discount_percent = 60

    discount_amount = price * (discount_percent / 100)
    final_price = price - discount_amount

    return final_price


# Test Cases
print("Test 1:", apply_discount(1000, 10))  # 10% discount
print("Test 2:", apply_discount(500))       # Default 5%
print("Test 3:", apply_discount(500,75)) # Discount more than 60%
```

## Task 2 factorial calculation

```python
# Recursive Factorial 
def factorial(n):
    if n < 0:
        print("Error: Negative number not allowed.")
        return None
    elif n ==0 or n==1:
        return 1
    else:
       factorial_output= n * factorial(n-1)
       return factorial_output
print("Factorial of 5 is:", factorial(5))
print("Factorial of 3 is:", factorial(3))
```

# Task 6

## Combined Utility Function

```python
# calculate discounted price and Filter the price which is more than 300
def process_prices(prices):
    discounted_prices = list(map(lambda price: price - (0.10*price), prices))
    
    discounted_price_above_300 = list(filter(lambda price: price > 300, discounted_prices))
    return discounted_prices , discounted_price_above_300

discounted_prices , discounted_price_above_300 = process_prices([100 ,500 ,900 ,50 ,750])
print("discounted_price", discounted_prices)
print(" Filter the prices above Rs.300",discounted_price_above_300)
```

## Mini problem using Menu Function

```python
# Task 7 - Menu Using Functions


# 1️ Add price
def add_price(prices_list, price):
    prices_list.append(price)
    print("Price added successfully!")


# 2 Get average price
def get_average_price(prices_list):
    if len(prices_list) == 0:
        return None

    total = 0
    for price in prices_list:
        total += price

    return total / len(prices_list)


# 3️ Get maximum price
def get_max_price(prices_list):
    if len(prices_list) == 0:
        return None

    maximum = prices_list[0]
    for price in prices_list:
        if price > maximum:
            maximum = price

    return maximum


#  Main Program
prices = []

while True:
    print("\n---- Price Menu ----")
    print("1 → Add price")
    print("2 → Show average price")
    print("3 → Show highest price")
    print("q → Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        try:
            price = float(input("Enter price: "))
            if price < 0:
                print("Price cannot be negative!")
            else:
                add_price(prices, price)
        except ValueError:
            print("Invalid input! Please enter a number.")

    elif choice == "2":
        avg = get_average_price(prices)

        if avg is None:
            print("No prices available.")
        else:
            print("Average Price:", round(avg, 2))

    elif choice == "3":
        highest = get_max_price(prices)

        if highest is None:
            print("No prices available.")
        else:
            print("Highest Price:", highest)

    elif choice == "q":
        print("Program exited.")
        break

    else:
        print("Invalid choice! Try again.")
```
